```python
#Adding another file of Country Code for analysis
dataset_country_code = pd.read_excel("/Users/kaveenaiyappa/Downloads/Zomatodataset/Country-Code.xlsx")
dataset_country_code.head()
```

```python
#Merging the two files
final_dataset=pd.merge(dataset,dataset_country_code,on="Country Code",how="left")
final_dataset
```

```python
#To check DataTypes present in the dataset
final_dataset.dtypes
```

```python
#Records for identifying the market presence of Zomato Internationally
country_names=final_dataset.Country.value_counts().index

country_values=final_dataset.Country.value_counts()

plt.pie(country_values[:3],labels=country_names[:3],autopct="%0.2f%%")
```

Zomato's Maximum Records or Transaction are in India. 

USA and UK take the following postions after India!

```python
#Rating Count will be the new column created by using the Groupbhy function on Aggregate Rating, Rating Color and Rating Text

ratings =final_dataset.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
```

# Conclusion
1. When rating is between 4.5 to 4.9, it is considered as excellent,

2. Between 4.0 to 4.4, the ratings are very good,

3. Between 3.5 to 3.9, the ratings are good,

4. Between 3.0 to 3.4, the ratings are average,

5. Between 2.5 to 2.9, it is Poor,

6. Below 2.5, it is considered as very Poor.

```python
#Plot the bar graph to visualize the ratings

import matplotlib
matplotlib.rcParams["figure.figsize"]=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color", data=ratings,palette=["white","red","orange","yellow","green","green"])
```

# Observation

Not Rated High Count is very high.


Maximum Number of Rating is between 2.5 to 3.4.
